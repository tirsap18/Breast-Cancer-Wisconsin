```python
# cek apakah terdapat outlier
df.plot(kind = "box", subplots =True, layout = (8,8), sharex = False, sharey = False, figsize=(20,25))
plt.show()
```

```python
df_satu = df
Q1 = df_satu.quantile(0.25)
Q3 = df_satu.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
df_out = df_satu[~((df_satu < (Q1 - 1.5 * IQR)) |(df_satu > (Q3 + 1.5 * IQR))).any(axis=1)]
```
